## Преобразование колонок

**EverBenched** и **LeaveOrNot** по сути - логические колонки

```python
transform_df = df.copy()

transform_df["LeaveOrNot"]
```

```python
transform_df["LeaveOrNot"] = transform_df["LeaveOrNot"].astype(bool)

transform_df
```

```python
transform_df["LeaveOrNot"].describe()
```

```python
def parse_ever_benched(value):
  return value == "Yes"

transform_df["EverBenched"] = transform_df["EverBenched"].apply(parse_ever_benched)

transform_df
```

```python
transform_df.query("EverBenched & not LeaveOrNot")
```

Давайте введем возрастные группы по десятилетиям (20-29, 30-39, 40-41 и тд)

```python
def make_age_bins(age):
  bin_names = ["twenties", "thirdties", "fourties"]
  bin_num = age // 10

  #2 тк минимум 20 лет
  bin_index = bin_num - 2

  return bin_names[bin_index]


transform_df["Age_bin"] = transform_df["Age"].apply(make_age_bins)

transform_df
```
